```python
# Concateno los 4 datasets para formar uno sólo llamado streaming_df
streaming_df = pd.concat([amz_df,dsy_df,hulu_df,ntfx_df])
streaming_df = streaming_df.reset_index(drop=True) 
```

```python
# Separo los datos de la columna duration colocando en una columna el numero y en otra la metrica del dato, minutos o temporadas
streaming_df['duration_num']=streaming_df['duration'].map(lambda x:x.split(' ')[0])
streaming_df['duration_type']=streaming_df['duration'].map(lambda x:x.split(' ')[1])
```

```python
streaming_df.head()
```

```python
# Agrego un indice a mi df
streaming_df['index'] = streaming_df.index + 1
```

```python
streaming_df.head()
```

```python
# Reordeno mis columnas
streaming_df = streaming_df[['index','type','title','cast','genero','release_year','duration','duration_num','duration_type','plataforma']]
```

```python
streaming_df.drop('duration', axis=1, inplace=True)
```

```python
# Cuento el total de valores duplicados
streaming_df.duplicated().value_counts()
```

```python
streaming_df.info()
```

```python
# Creo un nuevo df con los datos de cast y de index
actors_df = streaming_df[['index', 'cast']]
```

```python
# Separo los datos que se encuentran separados por coma 
actors_df['cast'] = actors_df['cast'].str.split(',')
```

```python
# Organizo la columna y queda un actor por cada fila
actors_df = actors_df.explode('cast')
```

```python
actors_df
```

```python
# Creo un nuevo df con los datos de genero y de index
genero_df =  streaming_df[['index', 'genero']]
```

```python
# Separo los datos que se encuentran separados por coma 
genero_df['genero'] = streaming_df['genero'].str.split(',')
```

```python
# Organizo la columna y queda un genero de pelicula por cada fila
genero_df = genero_df.explode('genero')
```

```python
genero_df
```

```python
# Se quitan los espacios para poder reemplazar el dato con exito
genero_df['genero'] = genero_df['genero'].str.strip()
```

```python
# Haciendo un tail encuentro que hay generos con nombres erroneos. Cambio comiedies por comedy
genero_df.replace({'Comedies':'Comedy'}, inplace=True)
```

```python
genero_df
```

```python
# Borro las columnas que saqué a parte cada una convertida en df
streaming_df.drop('cast', axis=1, inplace=True)
streaming_df.drop('genero', axis=1, inplace=True)
```
